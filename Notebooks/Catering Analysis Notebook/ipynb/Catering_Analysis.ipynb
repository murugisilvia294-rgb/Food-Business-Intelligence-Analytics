```python
import pandas as pd

Catering = pd.read_csv("Catering_Gigs.csv")
Catering.head()
```

Dataset Structure

```python
Catering.info()
```

Statistical Summary


```python
Catering.describe()
```

Total Revenue

```python
total_revenue = Catering['Price (KES)'].sum
print("Total Catering Revenue:",total_revenue)
```

Average Pack Value

```python
average_package = Catering['Price (KES)'].mean()
print("Average Package Value:",average_package)
```

Highest_Paying Tier

```python
highest_tier = Catering.loc[Catering['Price (KES)'].idxmax()]
print("Highest_Tier:", highest_tier)
```

Efficient Tier


```python
best_efficiency = Catering.loc[Catering['Price (KES)'].idxmin()]
print("Efficient Tier:", best_efficiency)
```

Cost Model

```python
Catering["Estimated Cost"] = 900
Catering.head()
```

Profit

```python
Catering["Profit"] = Catering['Price (KES)'] - Catering["Estimated Cost"]
Catering
```

Interactive Tier

```python
most_profitable = Catering.loc[Catering['Profit'].idxmax()]
print( most_profitable)
```

Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
plt.bar(Catering["Group Size"],Catering["Price (KES)"])

plt.title(" Revenue by Group Size")
plt.xlabel("Group Size")
plt.ylabel("Revenue(KES)")
plt.show()
```

Price Per Person Trend

```python
plt.figure(figsize=(8,5))
plt.plot(Catering["Group Size"],Catering["Price Per Person"],marker='o')

plt.title("Price Per Person Trend")
plt.xlabel("Group Size")
plt.ylabel("Price Per Person(KES)")

plt.show()
```

Profit By Group Size



```python
plt.figure(figsize=(8,5))
plt.bar(Catering["Group Size"],Catering["Profit"])

plt.title("Profit by Group Size")
plt.xlabel("Group Size")
plt.ylabel("Profit(KES)")

plt.show()
```
