```python
import pandas as pd

breakfast = pd.read_csv('Corporate_Breakfast.csv')
breakfast.head()
```

```python
breakfast.info()
```

Statistical Analysis

```python
breakfast.describe()
```

Total Revenue

```python
total_breakfast_revenue = breakfast['Price (KES)'].sum()
print("Total_Breakfast_Revenue:",total_breakfast_revenue)
```

Average Breakfast Price

```python
average_breakfast_price = breakfast["Price (KES)"].mean()
print("Average_Breakfast_Price:",average_breakfast_price)
```

Protein Dependency Analysis

```python
protein_counts = breakfast["Protein Type"].value_counts()
print("Protein_counts","\n",protein_counts)
```

Fruit Inclusion Analysis

```python
fruit_counts = breakfast["Fruit Included"].value_counts()
print("Fruit_counts:\n", fruit_counts)
```

Cost Simulation

```python
breakfast["Estimated Cost"]=120
```

Profit Calculation

```python
breakfast["Profit"]=breakfast["Price (KES)"]-breakfast["Estimated Cost"]
breakfast
```

Total Breakfast Profit

```python
total_breakfast_profit = breakfast["Profit"].sum()
print("Total_Breakfast_Profit:",total_breakfast_profit)
```

Visualization

Revenue By Day

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(breakfast["Day"], breakfast["Price (KES)"])
plt.xlabel("Day")
plt.ylabel("Revenue (KES)")
plt.title(" Breakfast Revenue by Day")
plt.show()
#
```

Protein Distribution

```python
plt.figure(figsize=(8, 5))
plt.bar(protein_counts.index, protein_counts.values)
plt.xlabel("Protein Type")
plt.ylabel("Count")
plt.title("Breakfast Protein Distribution")
plt.show()
#
#
```

Profit By Day

```python
plt.figure(figsize=(8, 5))
plt.bar(breakfast["Day"], breakfast["Profit"])
plt.xlabel("Day")
plt.ylabel("Profit (KES)")
plt.title("Breakfast Profit by Day")
plt.show()
#
```
