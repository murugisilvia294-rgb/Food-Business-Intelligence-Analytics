```python
import pandas as pd

Lunch= pd.read_csv("Corporate_Lunch.csv")
Lunch.head()
```

Structure

```python
Lunch.info()
```

Statistical Analysis

```python
Lunch.describe()
```

Total Lunch Revenue

```python
Total_Lunch_Revenue = Lunch['Price (KES)'].sum()
print("Total_Lunch_Revenue:",Total_Lunch_Revenue)

#
```

Average Lunch Price

```python
from numpy import average
average_lunch_price = Lunch['Price (KES)'].mean()
print("average_lunch_price:",average_lunch_price)
```

Highest Priced Meal

```python
highest_priced_meal = Lunch['Price (KES)'].max()
print("highest_priced_meal:",highest_priced_meal)
#
```

Protein Dependancy Analysis

```python
protein_dependancy = Lunch['Protein'].value_counts()
print("protein_dependancy:",protein_dependancy)
#
```

Cost Simulation

```python
Lunch["Estimated Cost"] = 150
```

Profit Calculation

```python
Lunch["Profit"]= Lunch['Price (KES)']-Lunch['Estimated Cost']
Lunch
```

Total Lunch Profit

```python
Total_Profit = Lunch['Profit'].sum()
print("Total_Profit:",Total_Profit)
#
```

Most Profitable Meal

```python
most_profitable_meal = Lunch.loc[Lunch['Profit'].idxmax()]
print("most_profitable_meal:",most_profitable_meal)
#
```

Revenue By Day

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(Lunch['Day'], Lunch['Price (KES)'])

plt.xlabel('Day')
plt.ylabel('Revenue (KES)')
plt.title('Lunch Revenue By Day')
plt.show()
#
```

Protein Distribution

```python
plt.figure(figsize=(8, 5))
plt.bar(protein_dependancy.index, protein_dependancy.values)

plt.xlabel('Protein')
plt.ylabel('count')
plt.title('Lunch Protein Distribution')
plt.show()
#
```

Profit By Day

```python
plt.figure(figsize=(8, 5))
plt.bar(Lunch['Day'], Lunch['Profit'])

plt.xlabel('Day')
plt.ylabel('Profit (KES)')
plt.title('Lunch Profit By Day')
plt.show()
#
```

Revenue vs Profit Comparison

```python
plt.figure(figsize=(8, 5))
x=range(len(Lunch['Day']))

plt.bar(x, Lunch['Price (KES)'], width=0.4, label='Revenue')
plt.bar(
    [i + 0.4 for i in x], Lunch['Profit'], width=0.4, label='Profit')
plt.xticks([i + 0.2 for i in x], Lunch['Day'])
plt.legend()

plt.xlabel('Day')
plt.ylabel('(KES)')
plt.title('Revenue vs Profit Comparison')
plt.show()
#
```
